Este notebook contiene los 4 WordClouds con las palabras que mas frecuencia tienen en los titulos de las plataformas. Hay uno para Coursera, EDX e UDEMY; y también agregue un ultimo WordCloud que contiene la unión de los 3 datasets.

IMPORTAMOS LIBRERIAS NECESARIAS

```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud,STOPWORDS
```

GENERAMOS UN CSV DE LOS TITULOS DE COURSERA PARA GENERAR EL WORD CLOUD

```python
wc_coursera = pd.read_csv('data_coursera_revisado.csv')
wc_edx = pd.read_csv('data_edx_revisado.csv')
wc_udemy = pd.read_csv('data_udemy_revisado.csv')
```

```python
coursera_titles = wc_coursera['name']
```

```python
coursera_titles.to_csv('wc_coursera_titles.csv')
```

```python
wc_1 = open('wc_coursera_titles.csv','r').read()
```

```python
wc_coursera = WordCloud(stopwords=STOPWORDS,min_word_length=4).generate(wc_1)
```

```python
plt.imshow(wc_coursera)
plt.axis('off')
plt.show()
```
